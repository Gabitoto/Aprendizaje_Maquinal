**Ejercicio 2:**
- Genera dos arrays con datos ficticios aleatorios: uno que represente pesos (kg) y otro alturas (m), redondee a 2 decimales.
- Calcula el Índice de Masa Corporal (IMC):  
$IMC = \frac{peso}{altura^2}$.

````python
pesos = np.random.randint(low=60, high=100, size=10)
alturas= np.random.randint(low=160, high=200, size=10)/100

print(f"el peso de cada humano es de: {pesos}")
print(f"la altura de cada humano es de: {alturas}")

count= 0
IMC = []
for a in pesos:
    count + 1
    resultado = (a/(alturas[count]**2))
    IMC.append(resultado.round(2))

print(IMC)

IMCNP = np.array(pesos/alturas**2).round(2)
print(IMCNP)
````

**Ejercicio 4:**

Realice los 3 gráficos anteriores en fila, uno al lado del otro usando subplots

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

# Cargar el dataset para los ejemplos
iris = load_iris()
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['class'])
df['class'] = df['class'].astype(int)
df['class'] = df['class'].replace({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Crear la figura y los subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# --- Gráfico 1: Barras (izquierda) ---
conteo_especies = df['class'].value_counts()
axes[0].bar(conteo_especies.index, conteo_especies.values, color=['skyblue', 'salmon', 'lightgreen'])
axes[0].set_title('Cantidad de Muestras por Especie', fontsize=14)
axes[0].set_xlabel('Especie', fontsize=12)
axes[0].set_ylabel('Cantidad de Muestras', fontsize=12)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
for index, value in enumerate(conteo_especies):
    axes[0].text(index, value, str(value), ha='center', va='bottom', fontsize=10)

# --- Gráfico 2: Histograma (centro) ---
# Accede a la columna usando el nombre exacto: 'petal length (cm)'
iris_petal = df["petal length (cm)"]
axes[1].hist(iris_petal, bins=7, color="skyblue", edgecolor="black", alpha=0.7)
axes[1].set_title('Histograma de Longitud de Pétalos', fontsize=14)
axes[1].set_xlabel('Longitud de Pétalo (cm)', fontsize=12)
axes[1].set_ylabel('Frecuencia', fontsize=12)
axes[1].grid(axis="y", linestyle="--", alpha=0.6)

# --- Gráfico 3: Dispersión (derecha) ---
sepal_length = iris.data[:, 0]
sepal_width = iris.data[:, 1]
axes[2].scatter(sepal_length, sepal_width, alpha=0.7)
axes[2].set_title("Relación Largo y Ancho del Sépalo", fontsize=14)
axes[2].set_xlabel("Largo del Sépalo (cm)", fontsize=12)
axes[2].set_ylabel("Ancho del Sépalo (cm)", fontsize=12)
axes[2].grid(True, linestyle="--", alpha=0.6)

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()

# Mostrar la figura
plt.show()
````

````python
# Creamos la grafica que vemos en la imagen
import pandas as pd
import matplotlib.pyplot as plt

# Fromula de pendiente: y = mx + b

# puntos a graficar:
a = (1,1)
b = (7,4)

# calculo de pendiente
m = ((b[1]-a[1])/(b[0]-a[0]))

# calculamos la intercepto
b = (a[0] - m*a[1])

print(f"La pendiente de la recta es: {m}")
print(f"La ecuación de la recta es: y = {m}x + {b}")

# Generar valores x para la recta
x = np.linspace(0, 8.5, 100) # Rango ligeramente extendido para que la línea se vea más allá de los puntos A y B
y = m * x + b

# Configurar el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'y = {m:.1f}x + {b:.1f}')

# Dibujar los puntos A y B
plt.scatter([1,7], [1,4], color='blue', zorder=5)
plt.text(1 + 0.2, 1 + 0.2, 'A', fontsize=12, verticalalignment='bottom')
plt.text(7 + 0.2, 4 + 0.2, 'B', fontsize=12, verticalalignment='bottom')

# Dibujar las líneas punteadas para 'rise' y 'run'
plt.plot([1, 7], [1, 1], 'k--', lw=1) # run
plt.plot([7, 7], [1, 4], 'k--', lw=1) # rise

# Añadir texto 'run' y 'rise'
plt.text((1 + 7) / 2, 1 - 0.3, 'run', fontsize=14, horizontalalignment='center')
plt.text(7 + 0.3, (1 + 4) / 2, 'rise', fontsize=14, verticalalignment='center', rotation=90)


# Título y etiquetas
plt.title('Slope = rise / run', fontsize=16)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14, rotation=0)

# Límites de los ejes para que coincidan con la imagen
plt.xlim(0, 8.5)
plt.ylim(0, 5.5)

# Cuadrícula
plt.grid(True)

# Mostrar el gráfico
plt.show()
````

Cargue los datos del archivo "datos.csv" de la carpeta `1_datos`, analice los valores de la columna `dataset` para separar los datos en 4 grupos y grafíquelos usando scatterplot y subplots (organice los gráficos en 2 filas y 2 columnas).

````python
datos = pd.read_csv("1_datos/datos.csv", delimiter = ",", header=0)
datus = pd.DataFrame(datos)
datus.head()
````

````python
import seaborn as sns

# Graficar
sns.scatterplot(datus,x='x',y='y', hue_order=["I","II","III","IV"],hue='dataset')
````

````python
# Crear la figura y los subplots (2 filas x 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Asegurar que los datasets se asignen a cada subplot
datasets = datus['dataset'].unique()
axes = axes.flatten()  # Para iterar fácilmente sobre los 4 ejes

# Suponiendo que tienes columnas 'x' e 'y' para graficar
for i, dataset in enumerate(datasets):
    subset = datus[datus['dataset'] == dataset]
    sns.scatterplot(
        data=subset,
        x="x", 
        y="y", 
        ax=axes[i],
        hue="dataset",  # Colorear por dataset
        palette="tab10",
        legend=False
    )
    axes[i].set_title(f"Dataset: {dataset}")

plt.tight_layout()
plt.show()
````
